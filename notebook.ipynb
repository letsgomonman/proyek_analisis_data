## Menentukan Pertanyaan Bisnis

**SMART Question** adalah sebuah framework untuk merumuskan pertanyaan secara terstruktur agar memeperoleh informasi yang mendalam.

* **Spesific (Spesifik)**
  * Pertanyaan harus jelas, fokus pada sebuah topik tertentu, dan tidak bermakna ganda. Hindari pertanyaan yang terlalu luas.
    * Salah: Bagaimana cara meningkatkan penjualan?
    * Benar: Faktor apa saja yang memengaruhi penurunan penjualan produk kategori elektronik di wilayah Jakarta selama kuartal terakhir?
* **Measurable (Terukur)**
  * Pertanyaan harus bisa dijawab dengan angka atau matrix yang konkret, harus tahu apa yang akan dihitung.
    * Salah: Apakah pelanggan senang dengan layanan kita?
    * Benar: Berapa skor rata-rata Customer Satisfaction untuk layanan purna jual bulan ini dibandingkan bulan lalu?
* **Action-Oriented (Berorientasi Aksi)**
  * Hasil dari pertanyaan harus bisa memberikan arahan untuk melakukan tindakan nyata. Jika pertanyaan terjawab, stakeholder harus tahu apa langkah selanjutnya.
    * Salah: Mengapa orang suka berbelanja?
    * Benar: Fitur apa pada aplikasi yang paling sering digunakan sebelum pengguna memutuskan untuk melakukan checkout?
* **Relevant (Relevan)**
  * Hasil dari pertanyaan harus sejalan dengan tujuan utama bisnis atau masalah yang sedang dihadapi.
    * Salah: Menanyakan tentang stok gudang saat masalah utamanya adalah efektivitas kampanye media sosial.
    * Benar: Apakah kampanye iklan di Instagram memberikan Return on Ad Spend (ROAS) yang lebih tinggi dibandingkan iklan di TikTok?
* **Time-bound (Terikat Waktu)**
  * Pertanyaan harus ada batasan waktu yang jelas agar analisis memiliki konteks yang tepat.
    * Salah: Berapa banyak pengguna baru kita?
    * Benar: Berapa tingkat pertumbuhan pengguna baru secara bulanan (Month-over-Month) sepanjang tahun 2025?

**Contoh pertanyaan bisnis yang memenuhi seluruh elemen SMART**

***"Faktor apa saja yang memengaruhi penurunan conversion rate pada pengguna aplikasi Android di wilayah Jabodetabek sebesar 5% selama periode Flash Sale Maret 2026?"***

Keterangan:

- **Specific**: Fokus pada "penurunan conversion rate" untuk "aplikasi Android" di "Jabodetabek". Bukan sekadar penjualan turun.
- **Measurable**: Ada angka konkret yang ingin dianalisis, yaitu penurunan sebesar "5%".
- **Action-Oriented**: Dengan mengetahui faktor penyebabnya misalnya bug pada versi Android tertentu atau kendala logistik di Jabodetabek, tim bisa langsung melakukan perbaikan teknis atau operasional.
- **Relevant**: Penurunan konversi saat Flash Sale adalah masalah kritis bagi bisnis retail/e-commerce.
- **Time-bound**: Dibatasi pada periode spesifik "Flash Sale Maret 2026".

- **Pertanyaan 1:** Apa saja kategori produk yang menghasilkan total pendapatan/revenue tertinggi dan terendah selama periode tahun 2018?

  - Specific: Fokus mencari kategori produk tertinggi dan terendah dari segi pendapatan
  - Measurable: Dapat diukur dengan angka konkret yaitu "total pendapatan".
  - Action-Oriented: Dapat memberikan arahan untuk bisnis memperbanyak stok pada produk yang paling laris atau memberikan diskon pada produk yang kurang laku.
  - Relevant: Diketahuinya produk yang paling menguntungkan sangat relevan dengan tujuan e-commerce.
  - Time-Bound: Dibatasi pada waktu yang jelas, yaitu "periode tahun 2018".
- **Pertanyaan 2:** Bagaimana  segmentasi pelanggan berdasarkan tingkat RFM (Recency, Frequency, Monetary) selama periode transaksi terakhir, dan kelompok pelanggan mana yang paling bernilai "Best Customers" untuk diberikan program loyalitas eksklusif?
  - Specific: Mencari kelompok pelanggan menggunakan metrik RFM.
  - Measurable: Perhitungan RFM secara konkret dari jumlah hari (Recency), jumlah total transaksi (Frequency), dan total uang yang dihabiskan (Monetary).
  - Action-Oriented: Memberi arahan kepada tim marketing untuk menargetkan "Best Customers" dengan program khusus.
  - Relevant: Mempertahankan pelanggan lama yang sering berbelanja yang merupakan strategi kunci di e-commerce.
  - Time-Bound: Diukur dengan batasan "periode transaksi terakhir"

## Visualization & Explanatory Analysis

### Pertanyaan 1: Apa saja kategori produk yang menghasilkan total pendapatan/revenue tertinggi dan terendah selama periode tahun 2018?

```python
all_df.sample(5)
```

```python
# Data khusus tahun 2018
year_2018_df = all_df[all_df['order_year'] == 2018]

year_2018_df.head()
```

```python
# Menghitung total pendapatan (price) per kategori produk
revenue_per_category_df = year_2018_df.groupby(by='product_category_name_english')['price'].sum().reset_index()

revenue_per_category_df.head()
```

```python
# Membuat Visualisasi
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(24, 6))

colors = ["#72BCD4", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3"]

# Grafik kiri 5 kategori pendapatan tertinggi
sns.barplot(
    x="price",
    y="product_category_name_english", data=revenue_per_category_df.head(5),
    palette=colors,
    ax=ax[0]
)
ax[0].set_ylabel(None)
ax[0].set_xlabel("Total Pendapatan", fontsize=12)
ax[0].set_title("Kategori Produk dengan Pendapatan Tertinggi (2018)", loc="center", fontsize=15)
ax[0].tick_params(axis='y', labelsize=12)

# Grafik kanan 5 kategori pendapatan terendah
sns.barplot(
    x="price",
    y="product_category_name_english",
    data=revenue_per_category_df.sort_values(by="price", ascending=True).head(5),
    palette=colors,
    ax=ax[1]
)
ax[1].set_ylabel(None)
ax[1].set_xlabel("Total Pendapatan", fontsize=12)
ax[1].invert_xaxis()
ax[1].yaxis.set_label_position("right")
ax[1].yaxis.tick_right()
ax[1].set_title("Kategori Produk dengan Pendapatan Terendah (2018)", loc="center", fontsize=15)

plt.suptitle("Performa Pendapatan Kategori Produk Tertinggi dan Terendah di Tahun 2018", fontsize=20)
plt.show()
```

### Pertanyaan 2: Bagaimana  segmentasi pelanggan berdasarkan tingkat RFM (Recency, Frequency, Monetary) selama periode transaksi terakhir, dan kelompok pelanggan mana yang paling bernilai "Best Customers" untuk diberikan program loyalitas eksklusif?

```python
# DataFrame untuk RFM
rfm_df = all_df.groupby(by="customer_unique_id", as_index=False).agg({
    "order_purchase_timestamp": "max",
    "order_id": "nunique",
    "price": "sum"
})

rfm_df.columns = ["customer_unique_id", "max_order_timestamp", "frequency", "monetary"]
```

```python
# Menghitung Recency
recent_date = all_df["order_purchase_timestamp"].max() + pd.Timedelta(days=1)

rfm_df["recency"] = rfm_df["max_order_timestamp"].apply(lambda x: (recent_date - x).days)

rfm_df.drop("max_order_timestamp", axis=1, inplace=True)

print(rfm_df.head())
```

```python
# Visualisasi RFM
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(30, 6))

colors = ["#72BCD4", "#72BCD4", "#72BCD4", "#72BCD4", "#72BCD4"]

# Grafik Recency
sns.barplot(y="recency", x="customer_unique_id", data=rfm_df.sort_values(by="recency", ascending=True).head(5), palette=colors, ax=ax[0])
ax[0].set_ylabel(None)
ax[0].set_xlabel("Customer ID (disingkat)", fontsize=12)
ax[0].set_title("By Recency (days)", loc="center", fontsize=18)
ax[0].tick_params(axis ='x', labelsize=10, rotation=45)
ax[0].set_xticklabels([val[:5]+'...' for val in rfm_df.sort_values(by="recency", ascending=True).head(5)['customer_unique_id']])

# Grafik Frequency
sns.barplot(y="frequency", x="customer_unique_id", data=rfm_df.sort_values(by="frequency", ascending=False).head(5), palette=colors, ax=ax[1])
ax[1].set_ylabel(None)
ax[1].set_xlabel("Customer ID (disingkat)", fontsize=12)
ax[1].set_title("By Frequency", loc="center", fontsize=18)
ax[1].tick_params(axis='x', labelsize=10, rotation=45)
ax[1].set_xticklabels([val[:5]+'...' for val in rfm_df.sort_values(by="frequency", ascending=False).head(5)['customer_unique_id']])

# Grafik Monetary
sns.barplot(y="monetary", x="customer_unique_id", data=rfm_df.sort_values(by="monetary", ascending=False).head(5), palette=colors, ax=ax[2])
ax[2].set_ylabel(None)
ax[2].set_xlabel("Customer ID (disingkat)", fontsize=12)
ax[2].set_title("By Monetary", loc="center", fontsize=18)
ax[2].tick_params(axis='x', labelsize=10, rotation=45)
ax[2].set_xticklabels([val[:5]+'...' for val in rfm_df.sort_values(by="monetary", ascending=False).head(5)['customer_unique_id']])

plt.suptitle("Best Customer Based on RFM Parameters (customer_unique_id)", fontsize=20)
plt.show()
```

## Analisis Lanjutan (Opsional)

```python
# Binning untuk Recency (jarak hari terakhir belanja)
bins_recency = [0, 30, 90, float('inf')]
labels_recency = ['Active (<= 30 hari)', 'Warm (31-90 hari)', 'Cold (> 90 hari)']
rfm_df['recency_segment'] = pd.cut(rfm_df['recency'], bins=bins_recency, labels=labels_recency)
```

```python
# Binning untuk Monetary (Total belanja) menggunakan quartil (qcut) agar terbagi rata
rfm_df['monetary_segment'] = pd.qcut(rfm_df['monetary'], q=3, labels=['Low', 'Medium', 'High'])
```

```python
# Jumlah pelanggan segmen recency
rfm_df['recency_segment'].value_counts()
```

```python
# Jumlah pelanggan segmen monetary
rfm_df['monetary_segment'].value_counts()
```

```python
# Hasil akhir tabel RFM yang sudah memiliki segmen
rfm_df[['customer_unique_id', 'recency', 'recency_segment', 'monetary', 'monetary_segment']].head()
```
